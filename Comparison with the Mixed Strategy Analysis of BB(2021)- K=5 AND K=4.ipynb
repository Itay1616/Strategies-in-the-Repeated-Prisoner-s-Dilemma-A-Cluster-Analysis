```python
def MLE_BB (data,AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB ,CSG0BB, CSGCCBB, CSGCDBB,CSGDCBB, CSGDDBB ,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB,CSG0BB2, CSGCCBB2, CSGCDBB2,CSGDCBB2, CSGDDBB2 ,CSG0BB3, CSGCCBB3, CSGCDBB3,CSGDCBB3, CSGDDBB3 ):
    for i in range(data.shape[0]):
       
        if data.iloc[i]['round'] == 1:
            if data.iloc[i]['coop'] == 'coop':
                data.loc[i, "AD"] = abs(ln(AD0BB))
                data.loc[i, "CSG"] = abs(ln(CSG0BB))
                data.loc[i, "DSG"] = abs(ln(DSG0BB))
                
                data.loc[i, "CSG2"] = abs(ln(CSG0BB2))
                data.loc[i, "CSG3"] = abs(ln(CSG0BB3))
                
                
            else:
                data.loc[i, "AD"] = abs(ln(1-AD0BB))
                data.loc[i, "CSG"] = abs(ln(1-CSG0BB))
                data.loc[i, "DSG"] = abs(ln(1-DSG0BB))
                data.loc[i, "CSG2"] = abs(ln(1-CSG0BB2))
                data.loc[i, "CSG3"] = abs(ln(1-CSG0BB3))
                
         
        
        else:
       
            if (data.iloc[i-1]['ocoop'] == "coop" and  data.iloc[i-1]['coop'] == "coop"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADCCBB))
                    data.loc[i, "CSG"] = abs(ln(CSGCCBB))
                    data.loc[i, "DSG"] = abs(ln(DSGCCBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(CSGCCBB))
                    data.loc[i, "CSG3"] = abs(ln(CSGCCBB))

                
                else:
                    data.loc[i, "AD"] = abs(ln(1-ADCCBB))
                    data.loc[i, "CSG"] = abs(ln(1-CSGCCBB))
                    data.loc[i, "DSG"] = abs(ln(1-DSGCCBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(1-CSGCCBB2))
                    data.loc[i, "CSG3"] = abs(ln(1-CSGCCBB3))
                    


            
            elif (data.iloc[i-1]['ocoop'] == "defect" and  data.iloc[i-1]['coop'] == "coop"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADCDBB))
                    data.loc[i, "CSG"] = abs(ln(CSGCDBB))
                    data.loc[i, "DSG"] = abs(ln(DSGCDBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(CSGCDBB2))
                    data.loc[i, "CSG3"] = abs(ln(CSGCDBB3))

                

                else:
                    data.loc[i, "AD"] = abs(ln(1-ADCDBB))
                    data.loc[i, "CSG"] = abs(ln(1-CSGCDBB))
                    data.loc[i, "DSG"] = abs(ln(1-DSGCDBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(1-CSGCDBB2))
                    data.loc[i, "CSG3"] = abs(ln(1-CSGCDBB3))

                    
            elif (data.iloc[i-1]['ocoop'] == "coop" and  data.iloc[i-1]['coop'] == "defect"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADDCBB))
                    data.loc[i, "CSG"] = abs(ln(CSGDCBB))
                    data.loc[i, "DSG"] = abs(ln(DSGDCBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(CSGDCBB2))
                    data.loc[i, "CSG3"] = abs(ln(CSGDCBB3))
                else:
                    data.loc[i, "AD"] = abs(ln(1-ADDCBB))
                    data.loc[i, "CSG"] = abs(ln(1-CSGDCBB))
                    data.loc[i, "DSG"] = abs(ln(1-DSGDCBB))
                    
                    data.loc[i, "CSG2"] = abs(ln(1-CSGDCBB2))
                    data.loc[i, "CSG3"] = abs(ln(1-CSGDCBB3))

                    
            else:
                if (data.iloc[i-1]['ocoop'] == "defect" and  data.iloc[i-1]['coop'] == "defect"):
                    if  data.iloc[i]['coop'] == "coop":
                        data.loc[i, "AD"] = abs(ln(ADDDBB))
                        data.loc[i, "CSG"] = abs(ln(CSGDDBB))
                        data.loc[i, "DSG"] = abs(ln(DSGDDBB))
                        
                        data.loc[i, "CSG2"] = abs(ln(CSGDDBB2))
                        data.loc[i, "CSG3"] = abs(ln(CSGDDBB3))

                    else:
                        data.loc[i, "AD"] = abs(ln(1-ADDDBB))
                        data.loc[i, "CSG"] = abs(ln(1-CSGDDBB))
                        data.loc[i, "DSG"] = abs(ln(1-DSGDDBB))
                                                  
                        data.loc[i, "CSG2"] = abs(ln(1-CSGDDBB2))
                        data.loc[i, "CSG3"] = abs(ln(1-CSGDDBB3))

                        
    return(data)

```

```python
def MLE_calculate_BB(data):
    table = pd.pivot_table(data, values=['AD',"CSG","DSG","CSG2","CSG3"], index=['ID'], aggfunc=np.sum)
    return (table)

def MLE_min_BB (data):
    mle = sum(data.min(axis=1))
    data['min'] = data.min(axis=1)
    return (mle,data)


def Cluster_BB (data):
    cluster = []
    for i in range (data.shape[0]):
        if data.iloc[i]['AD']== data.iloc[i]['min']:
            cluster.append(1)
        if data.iloc[i]['CSG']== data.iloc[i]['min']:
            cluster.append(2)
        if data.iloc[i]['DSG']== data.iloc[i]['min']:
            cluster.append(3)
        if data.iloc[i]['CSG2']== data.iloc[i]['min']:
            cluster.append(4)
        if data.iloc[i]['CSG3']== data.iloc[i]['min']:
            cluster.append(5)
            
            
    data['class'] = cluster
    return(data)


def vLookup_BB (data,table):
    data = pd.merge(data, table['class'], on ='ID', how ='inner',)
    return (data)


def N_class_BB (data):
    table2 = pd.pivot_table(data, values=["class"], index=['ID'], aggfunc=np.mean)
    
    class1 = sum(table2['class'] == 1)
    class2 = sum(table2['class'] == 2)
    class3 = sum(table2['class'] == 3)
    class4 = sum(table2['class'] == 4)
    class5 = sum(table2['class'] == 5)
    
    return(class1,class2,class3,class4,class5)

```

```python
def vector_BB (data):
    table2 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc=np.mean)
    
    AD0, ADCC, ADCD, ADDC, ADDD = table2.iloc[0,4],table2.iloc[0,0],table2.iloc[0,1],table2.iloc[0,2],table2.iloc[0,3]
    CSG0, CSGCC, CSGCD, CSGDC, CSGDD =table2.iloc[1,4],table2.iloc[1,0],table2.iloc[1,1],table2.iloc[1,2],table2.iloc[1,3]
    DSG0, DSGCC, DSGCD, DSGDC, DSGDD = table2.iloc[2,4],table2.iloc[2,0],table2.iloc[2,1],table2.iloc[2,2],table2.iloc[2,3]
    CSG02, CSGCC2, CSGCD2, CSGDC2, CSGDD2 =table2.iloc[3,4],table2.iloc[3,0],table2.iloc[3,1],table2.iloc[3,2],table2.iloc[3,3]
    CSG03, CSGCC3, CSGCD3, CSGDC3, CSGDD3 =table2.iloc[4,4],table2.iloc[4,0],table2.iloc[4,1],table2.iloc[4,2],table2.iloc[4,3]
    
    table3 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc='count')
    
    AD0c, ADCCc, ADCDc, ADDCc, ADDDc = table3.iloc[0,4],table3.iloc[0,0],table3.iloc[0,1],table3.iloc[0,2],table3.iloc[0,3]
    CSG0c, CSGCCc, CSGCDc, CSGDCc, CSGDDc =table3.iloc[1,4],table3.iloc[1,0],table3.iloc[1,1],table3.iloc[1,2],table3.iloc[1,3]
    DSG0c, DSGCCc, DSGCDc, DSGDCc, DSGDDc = table3.iloc[2,4],table3.iloc[2,0],table3.iloc[2,1],table3.iloc[2,2],table3.iloc[2,3]
    CSG0c2, CSGCCc2, CSGCDc2, CSGDCc2, CSGDDc2 =table3.iloc[3,4],table3.iloc[3,0],table3.iloc[3,1],table3.iloc[3,2],table3.iloc[3,3]
    CSG0c3, CSGCCc3, CSGCDc3, CSGDCc3, CSGDDc3 =table3.iloc[4,4],table3.iloc[4,0],table3.iloc[4,1],table3.iloc[4,2],table3.iloc[4,3]
    
    
    AD0 = (((AD0*AD0c)+(ADCC*ADCCc)+(ADCD*ADCDc)+(ADDC*ADDCc)+(ADDD*ADDDc))/(AD0c+ADCCc+ADCDc+ADDCc+ADDDc))
    ADCC = ADCD = ADDC = ADDD = AD0
    
    CSGCD = (((CSGCDc * CSGCD)+(CSGDCc * CSGDC))/(CSGDCc+CSGCDc))
    CSGDC = CSGCD
    
    DSGCD = (((DSGCDc * DSGCD)+(DSGDCc * DSGDC))/(DSGDCc+DSGCDc))
    DSGDC = DSGCD
    
    CSGCD2 = (((CSGCDc2 * CSGCD2)+(CSGDCc2 * CSGDC2))/(CSGDCc2+CSGCDc2))
    CSGDC2 = CSGCD2
    
    CSGCD3 = (((CSGCDc3 * CSGCD3)+(CSGDCc3 * CSGDC3))/(CSGDCc3+CSGCDc3))
    CSGDC3 = CSGCD3
    
    return(AD0, ADCC, ADCD, ADDC, ADDD,CSG0, CSGCC, CSGCD, CSGDC, CSGDD,DSG0, DSGCC, DSGCD, DSGDC, DSGDD ,CSG02, CSGCC2, CSGCD2, CSGDC2, CSGDD2,CSG03, CSGCC3, CSGCD3, CSGDC3, CSGDD3)
```

```python
def MLE_calculate_BB(data):
    table = pd.pivot_table(data, values=['AD',"CSG","DSG","CSG2"], index=['ID'], aggfunc=np.sum)
    return (table)

def MLE_min_BB (data):
    mle = sum(data.min(axis=1))
    data['min'] = data.min(axis=1)
    return (mle,data)


def Cluster_BB (data):
    cluster = []
    for i in range (data.shape[0]):
        if data.iloc[i]['AD']== data.iloc[i]['min']:
            cluster.append(1)
        if data.iloc[i]['CSG']== data.iloc[i]['min']:
            cluster.append(2)
        if data.iloc[i]['DSG']== data.iloc[i]['min']:
            cluster.append(3)
        if data.iloc[i]['CSG2']== data.iloc[i]['min']:
            cluster.append(4)

            
            
    data['class'] = cluster
    return(data)


def vLookup_BB (data,table):
    data = pd.merge(data, table['class'], on ='ID', how ='inner',)
    return (data)


def N_class_BB (data):
    table2 = pd.pivot_table(data, values=["class"], index=['ID'], aggfunc=np.mean)
    
    class1 = sum(table2['class'] == 1)
    class2 = sum(table2['class'] == 2)
    class3 = sum(table2['class'] == 3)
    class4 = sum(table2['class'] == 4)
    
    return(class1,class2,class3,class4)

```

```python
def vector_BB (data):
    table2 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc=np.mean)
    
    AD0, ADCC, ADCD, ADDC, ADDD = table2.iloc[0,4],table2.iloc[0,0],table2.iloc[0,1],table2.iloc[0,2],table2.iloc[0,3]
    CSG0, CSGCC, CSGCD, CSGDC, CSGDD =table2.iloc[1,4],table2.iloc[1,0],table2.iloc[1,1],table2.iloc[1,2],table2.iloc[1,3]
    DSG0, DSGCC, DSGCD, DSGDC, DSGDD = table2.iloc[2,4],table2.iloc[2,0],table2.iloc[2,1],table2.iloc[2,2],table2.iloc[2,3]
    CSG02, CSGCC2, CSGCD2, CSGDC2, CSGDD2 =table2.iloc[3,4],table2.iloc[3,0],table2.iloc[3,1],table2.iloc[3,2],table2.iloc[3,3]
   
    
    table3 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc='count')
    
    AD0c, ADCCc, ADCDc, ADDCc, ADDDc = table3.iloc[0,4],table3.iloc[0,0],table3.iloc[0,1],table3.iloc[0,2],table3.iloc[0,3]
    CSG0c, CSGCCc, CSGCDc, CSGDCc, CSGDDc =table3.iloc[1,4],table3.iloc[1,0],table3.iloc[1,1],table3.iloc[1,2],table3.iloc[1,3]
    DSG0c, DSGCCc, DSGCDc, DSGDCc, DSGDDc = table3.iloc[2,4],table3.iloc[2,0],table3.iloc[2,1],table3.iloc[2,2],table3.iloc[2,3]
    CSG0c2, CSGCCc2, CSGCDc2, CSGDCc2, CSGDDc2 =table3.iloc[3,4],table3.iloc[3,0],table3.iloc[3,1],table3.iloc[3,2],table3.iloc[3,3]
  
    
    AD0 = (((AD0*AD0c)+(ADCC*ADCCc)+(ADCD*ADCDc)+(ADDC*ADDCc)+(ADDD*ADDDc))/(AD0c+ADCCc+ADCDc+ADDCc+ADDDc))
    ADCC = ADCD = ADDC = ADDD = AD0
    
    CSGCD = (((CSGCDc * CSGCD)+(CSGDCc * CSGDC))/(CSGDCc+CSGCDc))
    CSGDC = CSGCD
    
    DSGCD = (((DSGCDc * DSGCD)+(DSGDCc * DSGDC))/(DSGDCc+DSGCDc))
    DSGDC = DSGCD
    
    CSGCD2 = (((CSGCDc2 * CSGCD2)+(CSGDCc2 * CSGDC2))/(CSGDCc2+CSGCDc2))
    CSGDC2 = CSGCD2

    
    return(AD0, ADCC, ADCD, ADDC, ADDD,CSG0, CSGCC, CSGCD, CSGDC, CSGDD,DSG0, DSGCC, DSGCD, DSGDC, DSGDD ,CSG02, CSGCC2, CSGCD2, CSGDC2, CSGDD2)
```

```python
def change_class (df1,df2,fit=0):
    for i in range (df2.shape[0]):
        if df2.iloc[i]['class']== df1.iloc[i]['class']:
            fit+=1
    return (fit/1734)
```

```python
def BB_Main_MLE (data,AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB ,CSG0BB, CSGCCBB, CSGCDBB, CSGDCBB, CSGDDBB
                 ,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB,CSG0BB2, CSGCCBB2, CSGCDBB2, CSGDCBB2, CSGDDBB2):
    
    BBclass1 = []
    BBclass2 = []
    BBclass3 = []
    BBclass4 = []
    BBMLE = []
    class_change = []
    
    for i in range (1) :
        
        if i >0 :
            BB = MLE_BB (data,AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB ,CSG0BB, CSGCCBB, CSGCDBB,CSGDCBB, CSGDDBB ,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB, CSG0BB2, CSGCCBB2, CSGCDBB2, CSGDCBB2, CSGDDBB2)
            table_BB = MLE_calculate_BB (BB)
            BBmle , table_BB = MLE_min_BB(table_BB)
            df_cluster_BB = Cluster_BB(table_BB)
            fitt = change_class(df_cluster_BB2,df_cluster_BB,fit=0)
            df_vlook_BB = vLookup_BB (BB, table_BB)
            class1BB,class2BB,class3BB, class4BB = N_class_BB (df_vlook_BB)
        
            
        else:
            BB = MLE_BB (data,AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB ,CSG0BB, CSGCCBB, CSGCDBB,CSGDCBB, CSGDDBB,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB, CSG0BB2, CSGCCBB2, CSGCDBB2, CSGDCBB2, CSGDDBB2)
            table_BB = MLE_calculate_BB (BB)
            BBmle , table_BB = MLE_min_BB(table_BB)
            df_cluster_BB = Cluster_BB(table_BB)
            df_vlook_BB = vLookup_BB (BB, table_BB)
            class1BB,class2BB,class3BB,class4BB = N_class_BB (df_vlook_BB)
            df_cluster_BB2 = df_cluster_BB
            fitt = 1
            
        
        AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB,CSG0BB, CSGCCBB, CSGCDBB, CSGDCBB, CSGDDBB,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB, CSG0BB2, CSGCCBB2, CSGCDBB2, CSGDCBB2, CSGDDBB2 = vector_BB(df_vlook_BB) 
          
        
        class_change.append(fitt)
        BBMLE.append(BBmle)
        BBclass1.append(class1BB)
        BBclass2.append(class2BB)
        BBclass3.append(class3BB)
        BBclass4.append(class4BB)
    
        
    
    
    return(BBMLE, BBclass1,BBclass2,BBclass3,BBclass4, class_change,AD0BB, ADCCBB, ADCDBB, ADDCBB, ADDDBB ,CSG0BB, CSGCCBB, CSGCDBB, CSGDCBB, CSGDDBB,DSG0BB, DSGCCBB, DSGCDBB, DSGDCBB, DSGDDBB,CSG0BB2, CSGCCBB2, CSGCDBB2, CSGDCBB2, CSGDDBB2)

```
