```python
AC50, AC5CC, AC5CD, AC5DC, AC5DD = 0.96,0.96,0.96,0.96,0.96
AD50, AD5CC, AD5CD, AD5DC, AD5DD = 0.04,0.04,0.04,0.04,0.04,
GRIM50, GRIM5CC, GRIM5CD, GRIM5DC, GRIM5DD= 0.91,0.91,0.09,0.09,0.09
STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD= 0.22,0.78,0.22,0.78,0.22
TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD = 0.9,0.9,0.1,0.9,0.1
```

```python
def MLE_df2018 (data,AC0, ACCC, ACCD, ACDC, ACDD,
            AD0, ADCC, ADCD, ADDC, ADDD,
            GRIM0, GRIMCC, GRIMCD, GRIMDC, GRIMDD,
            STFT0, STFTCC, STFTCD, STFTDC, STFTDD,
            TFT0, TFTCC, TFTCD, TFTDC, TFTDD):
    
    
    for i in range(data.shape[0]):
       
        if data.iloc[i]['round'] == 1:
            if data.iloc[i]['coop'] == 'coop':
                data.loc[i, "AC"] = abs(ln(AC0))
                data.loc[i, "AD"] = abs(ln(AD0))
                data.loc[i, "GRIM"] = abs(ln(GRIM0))
                data.loc[i, "STFT"] = abs(ln(STFT0))
                data.loc[i, "TFT"] = abs(ln(TFT0))

            else:
                data.loc[i, "AC"] = abs(ln(1-AC0))
                data.loc[i, "AD"] = abs(ln(1-AD0))
                data.loc[i, "GRIM"] = abs(ln(1-GRIM0))
                data.loc[i, "STFT"] = abs(ln(1-STFT0))
                data.loc[i, "TFT"] = abs(ln(1-TFT0))

         
        
        else:
       
            if (data.iloc[i-1]['ocoop'] == "coop" and  data.iloc[i-1]['coop'] == "coop"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADCC))
                    data.loc[i, "STFT"] = abs(ln(STFTCC))
                    data.loc[i, "GRIM"] = abs(ln(GRIMCC))
                    data.loc[i, "TFT"] = abs(ln(TFTCC))
                    data.loc[i, "AC"] = abs(ln(ACCC))
                else:
                    data.loc[i, "AD"] = abs(ln(1-ADCC))
                    data.loc[i, "STFT"] = abs(ln(1-STFTCC))
                    data.loc[i, "GRIM"] = abs(ln(1-GRIMCC))
                    data.loc[i, "TFT"] = abs(ln(1-TFTCC))
                    data.loc[i, "AC"] = abs(ln(1-ACCC))

            
            elif (data.iloc[i-1]['ocoop'] == "defect" and  data.iloc[i-1]['coop'] == "coop"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADCD))
                    data.loc[i, "STFT"] = abs(ln(STFTCD))
                    data.loc[i, "GRIM"] = abs(ln(GRIMCD))
                    data.loc[i, "TFT"] = abs(ln(TFTCD))
                    data.loc[i, "AC"] = abs(ln(ACCD))
                else:
                    data.loc[i, "AD"] = abs(ln(1-ADCD))
                    data.loc[i, "STFT"] = abs(ln(1-STFTCD))
                    data.loc[i, "GRIM"] = abs(ln(1-GRIMCD))
                    data.loc[i, "TFT"] = abs(ln(1-TFTCD))
                    data.loc[i, "AC"] = abs(ln(1-ACCD))

                    
            elif (data.iloc[i-1]['ocoop'] == "coop" and  data.iloc[i-1]['coop'] == "defect"):
                if  data.iloc[i]['coop'] == "coop":
                    data.loc[i, "AD"] = abs(ln(ADDC))
                    data.loc[i, "STFT"] = abs(ln(STFTDC))
                    data.loc[i, "GRIM"] = abs(ln(GRIMDC))
                    data.loc[i, "TFT"] = abs(ln(TFTDC))
                    data.loc[i, "AC"] = abs(ln(ACDC))
                else:
                    data.loc[i, "AD"] = abs(ln(1-ADDC))
                    data.loc[i, "STFT"] = abs(ln(1-STFTDC))
                    data.loc[i, "GRIM"] = abs(ln(1-GRIMDC))
                    data.loc[i, "TFT"] = abs(ln(1-TFTDC))
                    data.loc[i, "AC"] = abs(ln(1-ACDC))

                    
            else:
                if (data.iloc[i-1]['ocoop'] == "defect" and  data.iloc[i-1]['coop'] == "defect"):
                    if  data.iloc[i]['coop'] == "coop":
                        data.loc[i, "AD"] = abs(ln(ADDD))
                        data.loc[i, "STFT"] = abs(ln(STFTDD))
                        data.loc[i, "GRIM"] = abs(ln(GRIMDD))
                        data.loc[i, "TFT"] = abs(ln(TFTDD))
                        data.loc[i, "AC"] = abs(ln(ACDD))
                    else:
                        data.loc[i, "AD"] = abs(ln(1-ADDD))
                        data.loc[i, "STFT"] = abs(ln(1-STFTDD))
                        data.loc[i, "GRIM"] = abs(ln(1-GRIMDD))
                        data.loc[i, "TFT"] = abs(ln(1-TFTDD))
                        data.loc[i, "AC"] = abs(ln(1-ACDD))
                        
    return(data)

def MLE_calculate_df2018(data):
    table = pd.pivot_table(data, values=['AC','AD',"GRIM","STFT","TFT"], index=['ID'], aggfunc=np.sum)
    return (table)

def MLE_min_df2018 (data):
    mle = sum(data.min(axis=1))
    data['min'] = data.min(axis=1)
    return (mle,data)

def Cluster_df2018 (data):
    cluster = []
    for i in range (data.shape[0]):
        if data.iloc[i]['AC']== data.iloc[i]['min']:
            cluster.append(1)
        if data.iloc[i]['AD']== data.iloc[i]['min']:
            cluster.append(2)
        if data.iloc[i]['GRIM']== data.iloc[i]['min']:
            cluster.append(3)
        if data.iloc[i]['STFT']== data.iloc[i]['min']:
            cluster.append(4)
        if data.iloc[i]['TFT']== data.iloc[i]['min']:
            cluster.append(5)
            
    data['class'] = cluster
    return(data)

def change_class (df1,df2,fit=0):
    for i in range (df2.shape[0]):
        if df2.iloc[i]['class']== df1.iloc[i]['class']:
            fit+=1
    return (fit/1734)

def vLookup_df2018 (data,table):
    data = pd.merge(data, table['class'], on ='ID', how ='inner',)
    return (data)

def N_class_df2018 (data):
    table2 = pd.pivot_table(data, values=["class"], index=['ID'], aggfunc=np.mean)
    
    class1 = sum(table2['class'] == 1)
    class2 = sum(table2['class'] == 2)
    class3 = sum(table2['class'] == 3)
    class4 = sum(table2['class'] == 4)
    class5 = sum(table2['class'] == 5)
    
    return(class1,class2,class3,class4,class5)

def vector_df2018 (data):
    table2 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc=np.mean) 
    AC0, ACCC, ACCD, ACDC, ACDD = table2.iloc[0,4],table2.iloc[0,0],table2.iloc[0,1],table2.iloc[0,2],table2.iloc[0,3]
    AD0, ADCC, ADCD, ADDC, ADDD =table2.iloc[1,4],table2.iloc[1,0],table2.iloc[1,1],table2.iloc[1,2],table2.iloc[1,3]
    GRIM0, GRIMCC, GRIMCD, GRIMDC, GRIMDD = table2.iloc[2,4],table2.iloc[2,0],table2.iloc[2,1],table2.iloc[2,2],table2.iloc[2,3]
    STFT0, STFTCC, STFTCD, STFTDC, STFTDD = table2.iloc[3,4],table2.iloc[3,0],table2.iloc[3,1],table2.iloc[3,2],table2.iloc[3,3]
    TFT0, TFTCC, TFTCD, TFTDC, TFTDD = table2.iloc[4,4],table2.iloc[4,0],table2.iloc[4,1],table2.iloc[4,2],table2.iloc[4,3]
    
    table3 = pd.pivot_table(data, values=['roundone','aftercc','aftercd', 'afterdc', 'afterdd'],
                        index=["class"], aggfunc='count')
    
    AC0c, ACCCc, ACCDc, ACDCc, ACDDc = table3.iloc[0,4],table3.iloc[0,0],table3.iloc[0,1],table3.iloc[0,2],table3.iloc[0,3]
    AD0c, ADCCc, ADCDc, ADDCc, ADDDc =table3.iloc[1,4],table3.iloc[1,0],table3.iloc[1,1],table3.iloc[1,2],table3.iloc[1,3]
    GRIM0c, GRIMCCc, GRIMCDc, GRIMDCc, GRIMDDc = table3.iloc[2,4],table3.iloc[2,0],table3.iloc[2,1],table3.iloc[2,2],table3.iloc[2,3]
    STFT0c, STFTCCc, STFTCDc, STFTDCc, STFTDDc = table3.iloc[3,4],table3.iloc[3,0],table3.iloc[3,1],table3.iloc[3,2],table3.iloc[3,3]
    TFT0c, TFTCCc, TFTCDc, TFTDCc, TFTDDc = table3.iloc[4,4],table3.iloc[4,0],table3.iloc[4,1],table3.iloc[4,2],table3.iloc[4,3]
    
    AD0 = (((AD0*AD0c)+(ADCC*ADCCc)+(ADCD*ADCDc)+(ADDC*ADDCc)+(ADDD*ADDDc))/(AD0c+ADCCc+ADCDc+ADDCc+ADDDc))
    ADCC = ADCD = ADDC = ADDD = AD0
    
    AC0 = (((AC0*AC0c)+(ACCC*ACCCc)+(ACCD*ACCDc)+(ACDC*ACDCc)+(ACDD*ACDDc))/(AC0c+ACCCc+ACCDc+ACDCc+ACDDc))
    ACCC = ACCD = ACDC = ACDD = AC0
    
    GRIMDD = (((1-GRIM0)*GRIM0c)+((1-GRIMCC)*GRIMCCc)+(GRIMCD*GRIMCDc)+(GRIMDC*GRIMDCc)+(GRIMDD*GRIMDDc))/(GRIM0c+GRIMCCc+GRIMCDc+GRIMDCc+GRIMDDc)
    GRIMCD = GRIMDC = GRIMDD
    GRIMCC = GRIM0 = (1-GRIMDD)
    
    TFTDD = (((1-TFT0)*TFT0c)+((1-TFTCC)*TFTCCc)+(TFTCD*TFTCDc)+((1-TFTDC)*TFTDCc)+(TFTDD*TFTDDc))/(TFT0c+TFTCCc+TFTCDc+TFTDCc+TFTDDc)
    TFTCD = TFTDD
    TFTCC = TFT0 = TFTDC = (1-TFTDD)
    
    STFTDD = ((STFT0*STFT0c)+((1-STFTCC)*STFTCCc)+(STFTCD*STFTCDc)+((1-STFTDC)*STFTDCc)+(STFTDD*STFTDDc))/(STFT0c+STFTCCc+STFTCDc+STFTDCc+STFTDDc)
    STFTCD = STFT0 = STFTDD
    STFTCC  = STFTDC = (1-STFTDD)


    return(AC0, ACCC, ACCD, ACDC, ACDD,AD0, ADCC, ADCD, ADDC, ADDD
           ,GRIM0, GRIMCC, GRIMCD, GRIMDC, GRIMDD,
           STFT0, STFTCC, STFTCD, STFTDC, STFTDD,TFT0, TFTCC, TFTCD, TFTDC, TFTDD)
```

```python
def Main_MLE_df2018 (data,AC50,AC5CC, AC5CD, AC5DC, AC5DD,
               AD50, AD5CC, AD5CD, AD5DC, AD5DD,
               GRIM50, GRIM5CC,GRIM5CD, GRIM5DC, GRIM5DD,
               STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD,
               TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD):
    
    class115 = []
    class225 = []
    class335 = []
    class445 = []
    class555 = []
    fit = []
    MLE5 =[]
    
    
    for i in range (20) :
        
        if i >0 :
            
            # חישוב הנראות לכל וקטור
            df = MLE_df2018(data,AC50, AC5CC, AC5CD, AC5DC, AC5DD,
                        AD50, AD5CC, AD5CD, AD5DC, AD5DD,
                        GRIM50, GRIM5CC, GRIM5CD, GRIM5DC, GRIM5DD,
                        STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD,
                        TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD)

            # חישוב הפיבות תייבל לקבלת הסכום של הנראות לכל אסטרטגיה לכל שחקן
            table = MLE_calculate_df2018 (df)

            # חישוב הנראות לווקטור ושמירתו במשתנה והוספת עמודה שמכילה את המינמום לכל שחקן
            mle , table = MLE_min_df2018(table)

            # בחירת האסטרטגיה שהיא המינימום
            df_cluster = Cluster_df2018(table)
            
            # מציאת אחוז ההתאמה 
            fitt = change_class(df_cluster_2,df_cluster,fit=0)

            # איחוד לטבלה המרכזית את הקלאסר לכל שחקן
            df_vlook = vLookup_df2018 (df, table)

            # כמה שחקנים בכל קבוצה - בחינת השינויים לאורך זמן
            class15,class25,class35,class45,class55 = N_class_df2018 (df_vlook)

            
        else:
            
             # חישוב הנראות לכל וקטור
            df = MLE_df2018(data,AC50, AC5CC, AC5CD, AC5DC, AC5DD,
                        AD50, AD5CC, AD5CD, AD5DC, AD5DD,
                        GRIM50, GRIM5CC, GRIM5CD, GRIM5DC, GRIM5DD,
                        STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD,
                        TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD)

            # חישוב הפיבות תייבל לקבלת הסכום של הנראות לכל אסטרטגיה לכל שחקן
            table = MLE_calculate_df2018 (df)

            # חישוב הנראות לווקטור ושמירתו במשתנה והוספת עמודה שמכילה את המינמום לכל שחקן
            mle , table = MLE_min_df2018(table)

            # בחירת האסטרטגיה שהיא המינימום
            df_cluster = Cluster_df2018(table)
            
            df_cluster_2 = df_cluster
            
            # מציאת אחוז ההתאמה 
            fitt = 1

            # איחוד לטבלה המרכזית את הקלאסר לכל שחקן
            df_vlook = vLookup_df2018 (df, table)

            # כמה שחקנים בכל קבוצה - בחינת השינויים לאורך זמן
            class15,class25,class35,class45,class55 = N_class_df2018 (df_vlook)
            

        #הווקטור החדש
        AC50,AC5CC, AC5CD, AC5DC, AC5DD, AD50, AD5CC, AD5CD, AD5DC, AD5DD,GRIM50, GRIM5CC,GRIM5CD, GRIM5DC, GRIM5DD,STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD,TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD = vector_df2018(df_vlook)
        
        
        fit.append(fitt)
        MLE5.append(mle)
        class115.append(class15)
        class225.append(class25)
        class335.append(class35)
        class445.append(class45)
        class555.append(class55)

    
    return(MLE5, class115,class225,class335,class445,class555,fit,AC50, AC5CC, AC5CD, AC5DC, AC5DD,AD50, AD5CC, AD5CD, AD5DC, AD5DD, GRIM50, GRIM5CC,GRIM5CD, GRIM5DC,GRIM5DD, TFT50, TFT5CC, TFT5CD, TFT5DC, TFT5DD,STFT50, STFT5CC, STFT5CD, STFT5DC, STFT5DD)
```
